### Basic recommendation using only python
#### Basic understanding:
I first created a dictionary so I can quickly access any user's friends using their ID. Then, for each direct friend, I checked their friends. If the person was not me and not already my friend, I counted them as a mutual connection. Finally, I sorted users based on the number of mutual friends.

- load data from json file
- select the user and their direct friends
- and with individual direct friend we get the score of user
- and sort the score of recommended users desc
- finally getting the result as a list

```python
import json
def load_data(filename):
    with open(filename,'r') as f:
        return json.load(f)

def people_u_may_know(user_id, data):
    user_friends = {}
    for user in data['users']:
        user_friends[user['id']] = set(user['friends'])
        
    if user_id not in user_friends:
       return []
    direct_friends = user_friends[user_id]
    suggestions = {}
    for friends in direct_friends:
        for mutual in user_friends[friends]:
            if mutual != user_id and mutual not in direct_friends:
                suggestions[mutual] = suggestions.get(mutual, 0) + 1
    sorted_suggestion = sorted(suggestions.items(),key = lambda x:x[1], reverse=True)
    return [user_id for user_id, mutual_count in sorted_suggestion]
data = load_data('massive_data.json')
user_id = 3
recommendations = people_u_may_know(user_id, data)
print(recommendations)
```
